```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("housing.csv")
```

```python
data
```

```python
data.dropna(inplace = True)
data.info()
```

## Train test Split

```python
x = data.drop(['median_house_value'], axis = 1)
y = data['median_house_value']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x , y , test_size = 0.2)
```

```python
train_data = x_train.join(y_train)
train_data
```

## Data Exploration

```python
train_data.hist(figsize = (15 , 8))
```

```python
train_data.corr()
```

```python
plt.figure(figsize = (15 , 8))
sns.heatmap(train_data.corr() , annot = True , cmap = 'YlGnBu')
```

## Preprocesing

```python
train_data['total_rooms'] = np.log(train_data['total_rooms'] + 1)
train_data['total_bedrooms'] = np.log(train_data['total_bedrooms'] + 1)
train_data['population'] = np.log(train_data['population'] + 1)
train_data['households'] = np.log(train_data['households'] + 1)
#El +1 es para evitar valores en 0
#Log para que los datos tiendan mas a una distribuciÃ³n normal
```

```python
train_data.hist(figsize = (15,8))
```

```python
#Transformar la variable categorica
#Crearemos dummies para cada caracteristica
train_data.ocean_proximity.value_counts()
```

```python
train_data = train_data.join(pd.get_dummies(train_data.ocean_proximity)).drop(['ocean_proximity'] , axis = 1)
```

```python
plt.figure(figsize = (15 , 10))
sns.heatmap(train_data.corr() , annot = True , cmap = 'YlGnBu')
```

```python
plt.figure(figsize = (15,8))
sns.scatterplot(x="latitude" , y= "longitude", data = train_data , hue = "median_house_value")
#las casas que estan hacia la costa (abajo) tienden a ser mas costosas
```

## Feature engineering

```python
train_data['bedtoom_ratio'] = train_data['total_bedrooms']/train_data['total_rooms']
train_data['household_rooms'] = train_data['total_rooms']/train_data['households']
```

```python
plt.figure(figsize = (15 , 10))
sns.heatmap(train_data.corr() , annot = True , cmap = 'YlGnBu')
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
x_train, y_train = train_data.drop(['median_house_value'] , axis = 1), train_data['median_house_value']

reg = LinearRegression()

reg.fit(x_train , y_train)
```

```python
test_data = x_test.join(y_test)

test_data['total_rooms'] = np.log(test_data['total_rooms'] + 1)
test_data['total_bedrooms'] = np.log(test_data['total_bedrooms'] + 1)
test_data['population'] = np.log(test_data['population'] + 1)
test_data['households'] = np.log(test_data['households'] + 1)

test_data = test_data.join(pd.get_dummies(test_data.ocean_proximity)).drop(['ocean_proximity'] , axis = 1)

test_data['bedtoom_ratio'] = test_data['total_bedrooms']/test_data['total_rooms']
test_data['household_rooms'] = test_data['total_rooms']/test_data['households']


```

```python
x_test, y_test = test_data.drop(['median_house_value'] , axis = 1), test_data['median_house_value']
```

## Random Forest

```python
from sklearn.ensemble import RandomForestRegressor

forest = RandomForestRegressor()

forest.fit(x_train, y_train)
```

```python
forest.score(x_test, y_test)
```

### Grid search and cross validation

```python
from sklearn.model_selection import GridSearchCV

forest = RandomForestRegressor()

param_grid = {
    "n_estimators": [30, 50, 100],
    "max_features": [8, 12, 20]
    # "min_samples_split": [2, 4, 6, 8]

}

grid_search = GridSearchCV(forest, param_grid, cv = 5,
                            scoring = "neg_mean_squared_error", 
                            return_train_score = "True")


grid_search.fit(x_train, y_train)
```

```python
best_forest = grid_search.best_estimator_
```

```python
best_forest.score(x_test, y_test)
```
